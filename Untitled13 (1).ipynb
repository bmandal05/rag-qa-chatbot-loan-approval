```python
import kagglehub

path = kagglehub.dataset_download("sonalisingh1411/loan-approval-prediction")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(f"{path}/Training Dataset.csv")
df.fillna("N/A", inplace=True)
df.head()
```

```python
def row_to_text(row):
    return "\n".join([f"{col}: {val}" for col, val in row.items()])

docs = [row_to_text(row) for _, row in df.iterrows()]
```

```python
!pip install -q sentence-transformers faiss-cpu

from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

embedder = SentenceTransformer("all-MiniLM-L6-v2")

vectors = embedder.encode(docs, show_progress_bar=True)

index = faiss.IndexFlatL2(vectors.shape[1])
index.add(np.array(vectors))
```

```python
def retrieve_top_docs(query, k=3):
    query_vec = embedder.encode([query])
    _, indices = index.search(query_vec, k)
    return [docs[i] for i in indices[0]]
```

```python
!pip install -q transformers

from transformers import pipeline

generator = pipeline("text2text-generation", model="google/flan-t5-small")
```

```python
def rag_qa(query):
    retrieved = retrieve_top_docs(query, k=2)
    context = "\n\n".join(retrieved)

    prompt = f"""You are a helpful assistant and loan approval expert.

Context:
{context}

Answer the following question based on the above context:
Question: {query}
Answer:"""

    response = generator(prompt, max_new_tokens=150, do_sample=False)[0]['generated_text']
    return response
```

```python
question = "What are the reasons for loan rejection among applicants with low credit history?"
print("Answer:", rag_qa(question))
```
